```python
import pandas as pd
import numpy as np

train = pd.read_csv('Train.csv')
test = pd.read_csv('TestSet.csv')
# sample_sub = pd.read_csv('sample_submission.csv') # File not found
data_dict = pd.read_csv('data_dictionary.csv')
# sample_collection = pd.read_csv('sample_collection.csv') # File not found
target_pred_keep = pd.read_csv('TargetPred_To_Keep.csv')  # optional, see what it is

# Handle duplicate column names in 'train' DataFrame
cols = pd.Series(train.columns)
for dup in cols[cols.duplicated()].unique():
    cols[cols[cols == dup].index.values.tolist()] = [dup + '_' + str(i) if i != 0 else dup for i, i_val in enumerate(cols[cols == dup].index.values)]
train.columns = cols
```

```python
# Assume the first column is an ID (like 'ID' or 'SampleID')
id_col = train.columns[0]  # or explicitly 'ID'

# Define target columns (note: 'Na' was duplicated from previous cell, fixing it here)
target_cols = ['Al', 'B', 'Ca', 'Cu', 'Fe', 'K', 'Mg', 'Mn', 'N', 'Na', 'P', 'S', 'Zn']

# Get initial feature candidates from train, excluding ID and target columns
train_feature_candidates = [col for col in train.columns if col not in target_cols and col != id_col]

# Get feature candidates from test, excluding ID and 'Target_' columns
# The test set has 'Target_X' columns which are not features but represent what to predict.
test_feature_candidates = [col for col in test.columns if col != id_col and not col.startswith('Target_')]

# Determine the common feature columns present in both train and test that are actual features
common_feature_cols = list(set(train_feature_candidates) & set(test_feature_candidates))

# Sort for consistency (optional)
common_feature_cols.sort()

# Use the common features for both X_train and X_test
feature_cols = common_feature_cols

# Perform one-hot encoding for 'Depth_cm' on both train and test data
X_train = pd.get_dummies(train[feature_cols], columns=['Depth_cm'], drop_first=True)
X_test = pd.get_dummies(test[feature_cols], columns=['Depth_cm'], drop_first=True)

y_train = train[target_cols]
```

```python
from sklearn.metrics import mean_squared_error

def weighted_rmse(y_true, y_pred):
    # y_true and y_pred are DataFrames with same column order as target_cols
    weights = np.array([0.00,   # Al (not scored, but you still predict it)
                        0.05,   # B
                        0.10,   # Ca
                        0.10,   # Cu
                        0.10,   # Fe
                        0.10,   # K
                        0.10,   # Mg
                        0.10,   # Mn
                        0.10,   # N
                        0.10,   # Na
                        0.05,   # P
                        0.025,  # S
                        0.025]) # Zn

    # Ensure same order of columns
    y_true = y_true[target_cols]
    y_pred = y_pred[target_cols]

    rmses = []
    for i, col in enumerate(target_cols):
        rmse = np.sqrt(mean_squared_error(y_true[col], y_pred[col]))
        rmses.append(rmse)

    return np.sum(weights * rmses)
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer # Import SimpleImputer
import pandas as pd # Ensure pandas is imported for Series operations

models = {}
for col in target_cols:
    print(f"Training model for {col}...")

    # Impute missing values in the current target column for training
    y_tr_col = y_tr[col].copy() # Create a copy to avoid modifying original DataFrame directly

    if y_tr_col.isnull().any(): # Check if there are any NaN values
        imputer = SimpleImputer(strategy='mean')
        # Reshape for imputer: (n_samples,) -> (n_samples, 1)
        y_tr_col_imputed = imputer.fit_transform(y_tr_col.values.reshape(-1, 1))
        # Reshape back to (n_samples,) and convert to Series to keep index
        y_tr_col = pd.Series(y_tr_col_imputed.flatten(), index=y_tr_col.index)
        print(f"Imputed NaN values in target column: {col} using mean strategy.")

    rf = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
    rf.fit(X_tr, y_tr_col) # Use the imputed target column
    models[col] = rf
```

```python
# Make predictions on the validation set
predictions = pd.DataFrame(index=X_val.index)
for col in target_cols:
    model = models[col]
    pred = model.predict(X_val)
    predictions[col] = pred

# Calculate the weighted RMSE on the validation set
rmse_score = weighted_rmse(y_val, predictions)
print(f"Weighted RMSE on the validation set: {rmse_score}")
```

```python
# Make predictions on the test set
test_predictions = pd.DataFrame(index=X_test.index)
for col in target_cols:
    model = models[col]
    pred = model.predict(X_test)
    test_predictions[col] = pred

# Prepare the submission file
submission = pd.DataFrame({'ID': test['ID']})

# Rename columns to match submission format (e.g., 'Target_Al', 'Target_B')
submission_cols = []
for col in target_cols:
    submission_cols.append('Target_' + col)
test_predictions.columns = submission_cols

# Merge with the ID column
submission = pd.concat([submission, test_predictions], axis=1)

# Save the submission file
submission.to_csv('my_submission.csv', index=False)

# Now try to download the file
from google.colab import files
files.download('my_submission.csv')
```
